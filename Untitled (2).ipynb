```python
def calculate_RSI(data, window=14):
    delta = data['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    
    RS = gain / loss
    RSI = 100 - (100 / (1 + RS))
    
    return RSI

nifty_df['RSI'] = calculate_RSI(nifty_df)
banknifty_df['RSI'] = calculate_RSI(banknifty_df)

print("✅ RSI (Relative Strength Index) जोड़ दिया गया।")
```

```python
def calculate_MACD(data):
    data['EMA_12'] = data['Close'].ewm(span=12, adjust=False).mean()
    data['EMA_26'] = data['Close'].ewm(span=26, adjust=False).mean()
    data['MACD_Line'] = data['EMA_12'] - data['EMA_26']
    data['Signal_Line'] = data['MACD_Line'].ewm(span=9, adjust=False).mean()
    return data

nifty_df = calculate_MACD(nifty_df)
banknifty_df = calculate_MACD(banknifty_df)

print("✅ MACD Indicator जोड़ दिया गया।")
```

```python
def calculate_Bollinger_Bands(data, window=20):
    data['SMA_20'] = data['Close'].rolling(window=window).mean()
    data['Upper_Band'] = data['SMA_20'] + (data['Close'].rolling(window=window).std() * 2)
    data['Lower_Band'] = data['SMA_20'] - (data['Close'].rolling(window=window).std() * 2)
    return data

nifty_df = calculate_Bollinger_Bands(nifty_df)
banknifty_df = calculate_Bollinger_Bands(banknifty_df)

print("✅ Bollinger Bands जोड़ दिया गया।")
```

```python
def calculate_fibonacci_levels(df):
    high = df['High'].max()
    low = df['Low'].min()
    
    diff = high - low
    levels = {
        '0.0%': high,
        '23.6%': high - 0.236 * diff,
        '38.2%': high - 0.382 * diff,
        '50.0%': high - 0.5 * diff,
        '61.8%': high - 0.618 * diff,
        '78.6%': high - 0.786 * diff,
        '100.0%': low
    }
    
    return levels

nifty_fib_levels = calculate_fibonacci_levels(nifty_df)
banknifty_fib_levels = calculate_fibonacci_levels(banknifty_df)

print("NIFTY 50 Fibonacci Levels:\n", nifty_fib_levels)
print("\nBank NIFTY Fibonacci Levels:\n", banknifty_fib_levels)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

nifty_file = 'data/NIFTY_50.csv'
bank_nifty_file = 'data/BANK_NIFTY.csv'

nifty_data = pd.read_csv(nifty_file)
bank_nifty_data = pd.read_csv(bank_nifty_file)

nifty_data['Date'] = pd.to_datetime(nifty_data['Date'])
bank_nifty_data['Date'] = pd.to_datetime(bank_nifty_data['Date'])

nifty_X = nifty_data.drop(columns=['Close', 'Index Name', 'Date'])  
nifty_y = nifty_data['Close']  

bank_nifty_X = bank_nifty_data.drop(columns=['Close', 'Index Name', 'Date']) 
bank_nifty_y = bank_nifty_data['Close'] 

nifty_X = nifty_X.fillna(0)
bank_nifty_X = bank_nifty_X.fillna(0)

print(nifty_X.dtypes)
print(bank_nifty_X.dtypes)

nifty_X_train, nifty_X_test, nifty_y_train, nifty_y_test = train_test_split(nifty_X, nifty_y, test_size=0.2, random_state=42)
bank_X_train, bank_X_test, bank_y_train, bank_y_test = train_test_split(bank_nifty_X, bank_nifty_y, test_size=0.2, random_state=42)

nifty_model = LinearRegression()
bank_model = LinearRegression()

nifty_model.fit(nifty_X_train, nifty_y_train)
bank_model.fit(bank_X_train, bank_y_train)

print("मॉडल सफलतापूर्वक फिट हो गया।")
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

nifty_predictions = nifty_model.predict(nifty_X_test)
nifty_r2 = r2_score(nifty_y_test, nifty_predictions)
nifty_mae = mean_absolute_error(nifty_y_test, nifty_predictions)

print(f"NIFTY मॉडल R² स्कोर: {nifty_r2}")
print(f"NIFTY मॉडल MAE: {nifty_mae}")

bank_predictions = bank_model.predict(bank_X_test)
bank_r2 = r2_score(bank_y_test, bank_predictions)
bank_mae = mean_absolute_error(bank_y_test, bank_predictions)

print(f"बैंक NIFTY मॉडल R² स्कोर: {bank_r2}")
print(f"बैंक NIFTY मॉडल MAE: {bank_mae}")
```
